```python
def file_length(file):
    
    try:                          
        file = open(file)
        contents = file.read()
        file.close()
        print(len(contents))
    
    except:                       
        print(f"File name '{file}' not found.")
```

```python
class Marsupial:          
    def __init__(self,x,y):             
        self.pouch=[]                        
    def put_in_pouch(self,items): 
        self.pouch.append(items)   
    def pouch_contents(self):          
        return self.pouch

class Kangaroo(Marsupial):              
    def __init__(self,x,y):              
        super().__init__(x,y)          
        self.dx = 0                      
        self.dy = 0                        
    def jump(self,x,y):                 
        self.dx = self.dx + x                
        self.dy = self.dy + y
    def __str__(self):           
        return(f"I'm a Kangaroo located at the coordinates {self.dx},{self.dy}.")
```

```python
k = Kangaroo(0,0)
```

```python
print(k)
```

```python
k.put_in_pouch('doll')
```

```python
k.put_in_pouch('firetruck')
```

```python
k.put_in_pouch('kitten')
```

```python
 k.pouch_contents()
```

```python
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
arr=[]
for i in words:
    array=[]
    array.append(i.upper())
    array.append(i.lower())
    array.append(len(i))
    arr.append(array)
print(arr)
```

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
words = [word for word in words if len(word) >=4]
print(words)
```
